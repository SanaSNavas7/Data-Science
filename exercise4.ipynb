```python
import matplotlib.pyplot as plt

x, y = [], []
with open("data.txt", "r") as f:
    for line in f:
        values = line.strip().split() 
        x.append(float(values[0]))
        y.append(float(values[1]))


plt.plot(x, y, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot from data.txt")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]
plt.plot(x,y1,label="Even numbers",marker='o')
plt.plot(x,y2,label="Odd numbers",marker='s')
plt.title("MULTIPLE LINES EXAMPLE")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")  
plt.legend()  

plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 77, 6.7]


plt.bar(languages, popularity)
plt.title("Popularity of Programming Languages")
plt.show()


plt.barh(languages, popularity, color='red')
plt.title("Popularity of Programming Languages (Horizontal)")
plt.show()


colors = ['blue', 'green', 'orange', 'purple', 'red', 'cyan']
plt.bar(languages, popularity, color=colors)
plt.title("Popularity of Programming Languages (Colored)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(men))  
width = 0.35

plt.bar(x - width/2, men, width, label='Men', color='green')
plt.bar(x + width/2, women, width, label='Women', color='red')

plt.xlabel("Person")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")
plt.xticks(x, ['G1', 'G2', 'G3', 'G4', 'G5'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 77, 6.7]

plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title("Programming Languages Popularity")
plt.show()
```
