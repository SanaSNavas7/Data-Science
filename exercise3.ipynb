```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3])
y=np.array([4,5,6])
plt.plot(x,y,marker='s',color='blue',ms=10,mec='green',mfc='red')
plt.plot(x,y,'s--y')
plt.show()
```
